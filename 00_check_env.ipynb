Ejecuta esta celda para verificar OpenCV y NumPy.

```python
import cv2, numpy as np, matplotlib.pyplot as plt
print("OpenCV:", cv2.__version__)
print("NumPy:", np.__version__)

img = np.zeros((200, 300, 3), dtype=np.uint8)
cv2.rectangle(img, (40, 40), (260, 160), (0, 255, 0), 3)
cv2.putText(img, "Ok", (60, 120), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2, cv2.LINE_AA)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title("Imagen de Prueba")
plt.axis("off")
plt.show()
```
